```python
import os, glob, shutil

src_root = '/kaggle/working/Dataset Cleaned'
dst_root = '/kaggle/working/textile_combined'

os.makedirs(os.path.join(dst_root, 'bandhani'), exist_ok=True)
os.makedirs(os.path.join(dst_root, 'batik'), exist_ok=True)

# Copy Bandhani images
for f in glob.glob(os.path.join(src_root, 'bandhani', '*')):
    shutil.copy(f, os.path.join(dst_root, 'bandhani', os.path.basename(f)))

# Flatten Batik subclasses into a single folder
for root, dirs, files in os.walk(os.path.join(src_root, 'batik')):
    for filename in files:
        src = os.path.join(root, filename)
        dst = os.path.join(dst_root, 'batik', filename)
        shutil.copy(src, dst)

print("Bandhani count:", len(os.listdir(os.path.join(dst_root,'bandhani'))))
print("Batik count:", len(os.listdir(os.path.join(dst_root,'batik'))))
print("Combined dataset ready at:", dst_root)
```

```python
import os
from PIL import Image

input_dir = "/kaggle/working/Dataset Cleaned"
output_dir = "/kaggle/working/Dataset256"

os.makedirs(output_dir, exist_ok=True)

for folder in os.listdir(input_dir):
    sub_in = os.path.join(input_dir, folder)
    sub_out = os.path.join(output_dir, folder)
    os.makedirs(sub_out, exist_ok=True)

    for img_name in os.listdir(sub_in):
        try:
            img = Image.open(os.path.join(sub_in, img_name)).convert("RGB")
            img = img.resize((256, 256))
            img.save(os.path.join(sub_out, img_name))
        except:
            continue

print("Resizing complete!")
```
